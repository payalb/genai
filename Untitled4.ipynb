```python
from openai import OpenAI
import json
import os
import pprint
from IPython.display import Markdown
client = OpenAI( base_url="https://api.groq.com/openai/v1",
                     api_key="gsk_")

tool_create_directory = {
  "type": "function",
  "function": {
    "name": "create_directory",
    "description": "Creates a new directory with specified name",
    "parameters": {
      "type": "object",
      "properties": {
        "dir_name": {
          "type": "string",
          "description": "Name of the directory"
          }
        },
      "required": ["dir_name"]
    }
  }
}

def create_directory(dir_name):
        os.makedirs(dir_name, exist_ok=True)
        return f"Directory '{dir_name}' created successfully."

messages=[{"role": "user", "content": "Create a folder called 'my-folder'. Inside it create folder 'test-foler' "}]

def get_response():
  response = client.chat.completions.create(
  model="meta-llama/llama-4-scout-17b-16e-instruct",
  tools = [tool_create_directory],
  messages=messages
  )
  return response
  
  
response = get_response()
response_message = response.choices[0].message
tool_calls = response_message.tool_calls
print(tool_calls)
print(response_message)
if tool_calls:
        # Step 3: call the function
        # Note: the JSON response may not always be valid; be sure to handle errors
        available_functions = {
            "create_directory": create_directory
        }
        messages.append(response_message)
        # Step 4: send the info for each function call and function response to the model
        for tool_call in tool_calls:
            function_name = tool_call.function.name
            function_to_call = available_functions[function_name]
            function_args = json.loads(tool_call.function.arguments)
            function_response = function_to_call(
                dir_name=function_args.get("dir_name"),
            )
            messages.append(
                {
                    "tool_call_id": tool_call.id,
                    "role": "tool",
                    "name": function_name,
                    "content": str(function_response),
                }
            )
            pprint.pprint(messages)
second_response = client.chat.completions.create(
            model="meta-llama/llama-4-scout-17b-16e-instruct",
            messages=messages
)
print(second_response)
 
```
